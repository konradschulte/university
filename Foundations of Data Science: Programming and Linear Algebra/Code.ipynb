```python
###############################################################################################################
#####                                             MAIN WINDOW                                             #####
###############################################################################################################

# Main window of user interface
def main(): 
    
    # Define layout for main window    
    layout = [
        [sg.Text('Matrix Calculations', font=('bold'), justification='r')],
        [sg.Text('')],
        [sg.HorizontalSeparator()],
        [sg.Text('Please choose a Calculation Type')],
        [sg.HorizontalSeparator()],
        [sg.Text('')],
        [sg.Button('Addition & Subtraction', size = (35,1))],
        [sg.Text('')],
        [sg.Button('Multiplication', size = (35,1))],
        [sg.Text('')],
        [sg.Button('Determinant', size = (35,1))],
        [sg.Text('')],
        [sg.Button('Inverse', size = (35,1))],
        [sg.Text('')],
        [sg.Button('Eigenvalue & Eigenvector', size = (35,1))]
    ]
    
    # Define main window
    window = sg.Window('Written Product', layout, size=(400,430), element_justification='c', finalize=True)
    
    # Event loop
    while True:
        event, values = window.read()
        
        # Open a separate window for the respective calculation 
        if event == 'Multiplication':
            mult_window()
        if event == 'Addition & Subtraction':
            add_sub_window()
        if event == 'Determinant':
            det_window()
        if event == 'Inverse':
            inv_window()
        if event == 'Eigenvalue & Eigenvector':
            eig_window()
        
        # End if user closes window
        if event == sg.WIN_CLOSED:
            break
    
    window.close()
```

```python
###############################################################################################################
#####                                       MULTIPLICATION WINDOW                                         #####
###############################################################################################################

# Sub window for Multiplication       
def mult_window(): 
    
    # Define window for first input matrix (M1)
    matrix_1 = [  
        [sg.Text('Please indicate Dimensions of Matrix 1:')],
        [
            sg.Text('Rows:'),
            sg.Slider((1,10), enable_events=True, default_value=1, \
                      size=(12,20), key='-row_1-', orientation='h'),
            sg.Text('Columns:'),
            sg.Slider((1,10), enable_events=True, default_value=1, \
                      size=(12,20), key='-col_1-', orientation='h')
        ],
        [sg.HorizontalSeparator()],
        [sg.Text('Please fill in the respective values for Matrix 1:')]
    ]
    # Add 10x10 input boxes to M1
    for i in range(10):
        box_list=[]
        for j in range(10):
            # (First box should be the only one visible from the beginning)
            if i == 0 and j == 0: # 
                box_list.append(sg.pin(sg.In(size=(3,2), enable_events=True, \
                key = '1_'+str(i+1)+'_'+str(j+1), visible=True, justification='c')))
            else:
                box_list.append(sg.pin(sg.In(size=(3,2), enable_events=True, \
                key = '1_'+str(i+1)+'_'+str(j+1), visible=False, justification='c')))            
        matrix_1.append(box_list) 
    
    # Define window for second input matrix (M2)
    matrix_2 = [
        [sg.Text('Please indicate Dimensions of Matrix 2:')],
        [
            sg.Text('Rows:'),
            sg.Slider((1,10), enable_events=True, default_value=1, \
                      size=(12,20), key='-row_2-', orientation='h'),
            sg.Text('Columns:'),
            sg.Slider((1,10), enable_events=True, default_value=1, \
                      size=(12,20), key='-col_2-', orientation='h')
        ],
        [sg.HorizontalSeparator()],
        [sg.Text('Please fill in the respective values for Matrix 2:')]
    ]
    # Add 10x10 input boxes to M2
    for i in range(10):
        box_list=[]
        for j in range(10):
            # (First box should be the only one visible from the beginning)
            if i == 0 and j == 0: # 
                box_list.append(sg.pin(sg.In(size=(3,2), enable_events=True, \
                key = '2_'+str(i+1)+'_'+str(j+1), visible=True, justification='c')))
            else:
                box_list.append(sg.pin(sg.In(size=(3,2), enable_events=True, \
                key = '2_'+str(i+1)+'_'+str(j+1), visible=False, justification='c')))            
        matrix_2.append(box_list) 
    
    # Define window for output matrix (M3)
    matrix_3 = []
    # Add 10x10 input boxes to M3
    for i in range(10):
        box_list=[]
        for j in range(10):
            # (In the beginning, all boxes should be invisible)
            box_list.append(sg.pin(sg.In(size=(8,2), enable_events=True, \
            key = '3_'+str(i+1)+'_'+str(j+1), visible=False, justification='c')))            
        matrix_3.append(box_list) 
        
    # Define layout for multiplication window
    layout_mult = [[
                sg.Column(matrix_1, size=(360, 380), element_justification='c'),
                sg.VSeparator(),
                sg.Column(matrix_2, size=(360, 380), element_justification='c')
    ],
                [sg.HorizontalSeparator()],
                [sg.Button('Compute Multiplication', key= '-mult-', size = (35,1), visible=False)],
                [sg.Text('The resulting Matrix is:', key = '-answer-',visible=False)],
                [sg.Column(matrix_3)],
                [sg.Text('Matrix 2 must have the same number of rows as the number of columns of Matrix 1!', \
                         key = '-E1-', text_color='red', background_color='white',visible=False)],
                [sg.Text('Please fill in all missing values!', key = '-E2-',text_color='red', \
                         background_color='white',visible=True)],    
                [sg.Text('All Entries must be floats or integers!', key = '-E3-',text_color='red', \
                         background_color='white',visible=True)],    
    ]
    
    # Define multiplication window
    window_mult = sg.Window('Matrix Calculations: Multiplication', layout_mult, \
                            size=(780,740), element_justification='c', modal=True)
    
    # Event loop
    while True:
        event, values = window_mult.read()
        
        # Create (show) respective input fields for M1 rows
        if event == '-row_1-':
            for i in range(int(values[event])):
                for j in range(int(values['-col_1-'])):
                    window_mult['1_'+str(i+1)+'_'+str(j+1)].Update(visible = True)
                    for k in range(i+1,10):
                        window_mult['1_'+str(k+1)+'_'+str(j+1)].Update(visible = False)
            
            # Hide answer and matrix when changing number of rows
            window_mult['-answer-'].Update(visible=False)
            for i in range(10):
                for j in range(10):
                    window_mult['3_'+str(i+1)+'_'+str(j+1)].Update(visible = False)
                            
        # Create (show) respective input fields for M1 columns    
        if event == '-col_1-':
            for i in range(int(values[event])):
                for j in range(int(values['-row_1-'])):
                    window_mult['1_'+str(j+1)+'_'+str(i+1)].Update(visible = True)
                    for k in range(i+1,10):
                        window_mult['1_'+str(j+1)+'_'+str(k+1)].Update(visible = False)
            
            # Hide answer and matrix when changing number of columns
            window_mult['-answer-'].Update(visible=False)
            for i in range(10):
                for j in range(10):
                    window_mult['3_'+str(i+1)+'_'+str(j+1)].Update(visible = False)
                        
        # Create (show) respective input fields for M2 rows
        if event == '-row_2-':
            for i in range(int(values[event])):
                for j in range(int(values['-col_2-'])):
                    window_mult['2_'+str(i+1)+'_'+str(j+1)].Update(visible = True)
                    for k in range(i+1,10):
                        window_mult['2_'+str(k+1)+'_'+str(j+1)].Update(visible = False)
            
            # Hide answer and matrix when changing number of rows
            window_mult['-answer-'].Update(visible=False)
            for i in range(10):
                for j in range(10):
                    window_mult['3_'+str(i+1)+'_'+str(j+1)].Update(visible = False)
                            
        # Create (show) respective input fields for M2 columns    
        if event == '-col_2-':
            for i in range(int(values[event])):
                for j in range(int(values['-row_2-'])):
                    window_mult['2_'+str(j+1)+'_'+str(i+1)].Update(visible = True)
                    for k in range(i+1,10):
                        window_mult['2_'+str(j+1)+'_'+str(k+1)].Update(visible = False)
            
            # Hide answer and matrix when changing number of columns
            window_mult['-answer-'].Update(visible=False)
            for i in range(10):
                for j in range(10):
                    window_mult['3_'+str(i+1)+'_'+str(j+1)].Update(visible = False)
            
        # Show error message if columns of M1 and rows of M2 are not matching (necessity for multiplication)
        try:
            if values['-row_2-'] != values['-col_1-']:
                window_mult['-E1-'].Update(visible = True)
                for i in range(10):
                    for j in range(10):
                        window_mult['3_'+str(i+1)+'_'+str(j+1)].Update(visible = False)
                window_mult['-answer-'].Update(visible=False) 
            else:
                window_mult['-E1-'].Update(visible = False)
        except:
            pass
        
        # Show error message if not all values are filled in
        try:
            check_fill = 0
            for i in range(int(values['-row_1-'])):
                for j in range(int(values['-col_1-'])):
                    if values['1_'+str(i+1)+'_'+str(j+1)] == '':
                        check_fill += 1
            for k in range(int(values['-row_2-'])): # col_1 = row_2
                for l in range(int(values['-col_2-'])):
                    if values['2_'+str(k+1)+'_'+str(l+1)] == '':
                        check_fill += 1
            if check_fill > 0:
                window_mult['-E2-'].Update(visible = True)
                for i in range(10):
                    for j in range(10):
                        window_mult['3_'+str(i+1)+'_'+str(j+1)].Update(visible = False)
                window_mult['-answer-'].Update(visible=False)  
            else:
                window_mult['-E2-'].Update(visible = False)
        except:
            pass

        # Show error message if not all values are floats/integers
        try:
            check_str = 0
            for i in range(int(values['-row_1-'])):
                for j in range(int(values['-col_1-'])):
                    if isfloat(values['1_'+str(i+1)+'_'+str(j+1)]) == False:
                        check_str += 1
            for k in range(int(values['-row_2-'])):
                for l in range(int(values['-col_2-'])):
                    if isfloat(values['2_'+str(k+1)+'_'+str(l+1)]) == False:
                        check_str += 1
            if check_str > 0:
                window_mult['-E3-'].Update(visible = True)
                for i in range(10):
                    for j in range(10):
                        window_mult['3_'+str(i+1)+'_'+str(j+1)].Update(visible = False)
                window_mult['-answer-'].Update(visible=False)  
            else:
                window_mult['-E3-'].Update(visible = False)
        except:
            pass
                            
        # Show multiplication button if columns and rows match & all values are filled in
        try:
            if values['-row_2-'] == values['-col_1-'] and check_fill == 0 and check_str == 0:
                window_mult['-mult-'].Update(visible = True)
            else:
                window_mult['-mult-'].Update(visible = False)
        except:
            pass
    
        
        # Compute multiplication
        if event == '-mult-':
            
            # Hide answer and matrix (to prevent user from seeing too many boxes 
            # when computing a new multiplication with less rows and columns)
            for i in range(10):
                    for j in range(10):
                        window_mult['3_'+str(i+1)+'_'+str(j+1)].Update(visible = False)
            
            # Transfer values from first input matrix to a list (of lists)
            M_1 = []
            for i in range(int(values['-row_1-'])):
                row = []
                for j in range(int(values['-col_1-'])):
                    while True:
                        try:
                            row.append(float(values['1_'+str(i+1)+'_'+str(j+1)]))
                            break
                        except:
                            continue
                M_1.append(row)
                
            # Transfer values from second input matrix to a list (of lists)
            M_2 = []
            for i in range(int(values['-row_2-'])):
                row = []
                for j in range(int(values['-col_2-'])):
                    while True:
                        try:
                            row.append(float(values['2_'+str(i+1)+'_'+str(j+1)]))
                            break
                        except:
                            continue
                M_2.append(row)
                
            # Calculate resulting matrix
            M_3 = [] 
            # Defines how many rows the resulting matrix has 
            for i in range(int(values['-row_1-'])): 
                row = []
                # Defines how many columns the resulting matrix has
                for j in range(int(values['-col_2-'])): 
                    ent = 0
                    # Calculates all indices of M3 rowwise (one index equals the sum of k multi-  
                    # plications of all indices of a specific M1 row and a specific M2 column)
                    for k in range(int(values['-col_1-'])): # C_1 = R_2
                        ent += M_1[i][k]*M_2[k][j]
                    row.append(ent)
                M_3.append(row)
                        
            # Show results
            for i in range(int(values['-row_1-'])):
                for j in range(int(values['-col_2-'])):
                    window_mult['3_'+str(i+1)+'_'+str(j+1)].Update(visible=True)
                    window_mult['3_'+str(i+1)+'_'+str(j+1)].Update(str(M_3[i][j])) 
            window_mult['-answer-'].Update(visible=True)  
            
        # End if user closes window (jump back to main window)
        if event == sg.WIN_CLOSED:
            break
    
    window_mult.close()
    
#if __name__ == "__main__":
#    mult_window()
```

```python
###############################################################################################################
#####                                     ADDITION/SUBTRACTION WINDOW                                     #####
###############################################################################################################

# Sub window for Addition/Subtraction       
def add_sub_window(): 
    
    # Define window for first input matrix
    matrix_1 = [
        [sg.Text('Please indicate Dimensions of Matrix 1:')],
        [
            sg.Text('Rows:'),
            sg.Slider((1,10), enable_events=True, default_value=1, \
                      size=(12,20), key='-row_1-', orientation='h'),
            sg.Text('Columns:'),
            sg.Slider((1,10), enable_events=True, default_value=1, \
                      size=(12,20), key='-col_1-', orientation='h')
        ],
        [sg.HorizontalSeparator()],
        [sg.Text('Please fill in the respective values for Matrix 1:')],
    ]
    # Add 10x10 input boxes to first input matrix 
    for i in range(10):
        box_list=[]
        for j in range(10):
            # (First box should be the only one visible from the beginning)
            if i == 0 and j == 0: # 
                box_list.append(sg.pin(sg.In(size=(3,2), enable_events=True, \
                key = '1_'+str(i+1)+'_'+str(j+1), visible=True, justification='c')))
            else:
                box_list.append(sg.pin(sg.In(size=(3,2), enable_events=True, \
                key = '1_'+str(i+1)+'_'+str(j+1), visible=False, justification='c')))            
        matrix_1.append(box_list) 
        
    # Define window for second input matrix    
    matrix_2 = [
        [sg.Text('Please indicate Dimensions of Matrix 2:')],
        [
            sg.Text('Rows:'),
            sg.Slider((1,10), enable_events=True, default_value=1, \
                      size=(12,20), key='-row_2-', orientation='h'),
            sg.Text('Columns:'),
            sg.Slider((1,10), enable_events=True, default_value=1, \
                      size=(12,20), key='-col_2-', orientation='h')
        ],
        [sg.HorizontalSeparator()],
        [sg.Text('Please fill in the respective values for Matrix 2:')],
    ]
    # Add 10x10 input boxes to first input matrix 
    for i in range(10):
        box_list=[]
        for j in range(10):
            # (First box should be the only one visible from the beginning)
            if i == 0 and j == 0: # 
                box_list.append(sg.pin(sg.In(size=(3,2), enable_events=True, \
                key = '2_'+str(i+1)+'_'+str(j+1), visible=True, justification='c')))
            else:
                box_list.append(sg.pin(sg.In(size=(3,2), enable_events=True, \
                key = '2_'+str(i+1)+'_'+str(j+1), visible=False, justification='c')))            
        matrix_2.append(box_list)
        
    # Define window for output matrix    
    matrix_3 = []
    # Add 10x10 input boxes to output matrix 
    for i in range(10):
        box_list=[]
        for j in range(10):
            # (In the beginning, all boxes should be invisible)
            box_list.append(sg.pin(sg.In(size=(8,2), enable_events=True, \
            key = '3_'+str(i+1)+'_'+str(j+1), visible=False, justification='c')))            
        matrix_3.append(box_list)        
        
    # Define layout for addition/subtraction window
    layout_add_sub = [[
                sg.Column(matrix_1, size=(360, 380), element_justification='c'),
                sg.VSeparator(),
                sg.Column(matrix_2, size=(360, 380), element_justification='c')
    ],
                [sg.HorizontalSeparator()],
                [sg.pin(sg.Button('Compute Addition', key= '-add-', size = (35,1), visible=False)),
                 sg.pin(sg.Button('Compute Subtraction', key= '-sub-', size = (35,1), visible=False))],
                [sg.Text('The resulting Matrix is:', key = '-answer-',visible=False)],
                [sg.Column(matrix_3)],
                [sg.Text('Matrix 1 and Matrix 2 must have the same number of Rows and Columns!', \
                         key = '-E1-', text_color='red', background_color='white',visible=False)],
                [sg.Text('Please fill in all missing values!', key = '-E2-',text_color='red', \
                         background_color='white',visible=True)],    
                [sg.Text('All Entries must be floats or integers!', key = '-E3-',text_color='red', \
                         background_color='white',visible=True)],    
    ]
    
    # Define addition/subtraction window
    window_add_sub = sg.Window('Matrix Calculations: Addition & Subtraction', layout_add_sub, \
                               size=(780,740), element_justification='c', modal=True)
    
    # Event loop
    while True:
        event, values = window_add_sub.read()
            
        # Create (show) respective input fields for M1 rows
        if event == '-row_1-':
            for i in range(int(values[event])):
                for j in range(int(values['-col_1-'])):
                    window_add_sub['1_'+str(i+1)+'_'+str(j+1)].Update(visible = True)
                    for k in range(i+1,10):
                        window_add_sub['1_'+str(k+1)+'_'+str(j+1)].Update(visible = False)
            
            # Hide answer and matrix when changing number of rows
            window_add_sub['-answer-'].Update(visible=False)
            for i in range(10):
                for j in range(10):
                    window_add_sub['3_'+str(i+1)+'_'+str(j+1)].Update(visible = False)
                            
        # Create (show) respective input fields for M1 columns   
        if event == '-col_1-':
            for i in range(int(values[event])):
                for j in range(int(values['-row_1-'])):
                    window_add_sub['1_'+str(j+1)+'_'+str(i+1)].Update(visible = True)
                    for k in range(i+1,10):
                        window_add_sub['1_'+str(j+1)+'_'+str(k+1)].Update(visible = False)
            
            # Hide answer and matrix when changing number of columns
            window_add_sub['-answer-'].Update(visible=False)
            for i in range(10):
                for j in range(10):
                    window_add_sub['3_'+str(i+1)+'_'+str(j+1)].Update(visible = False)
                        
        # Create (show) respective input fields for M2 rows
        if event == '-row_2-':
            for i in range(int(values[event])):
                for j in range(int(values['-col_2-'])):
                    window_add_sub['2_'+str(i+1)+'_'+str(j+1)].Update(visible = True)
                    for k in range(i+1,10):
                        window_add_sub['2_'+str(k+1)+'_'+str(j+1)].Update(visible = False)
            
            # Hide answer and matrix when changing number of rows
            window_add_sub['-answer-'].Update(visible=False)
            for i in range(10):
                for j in range(10):
                    window_add_sub['3_'+str(i+1)+'_'+str(j+1)].Update(visible = False)
                            
        # Create (show) respective input fields for M2 columns   
        if event == '-col_2-':
            for i in range(int(values[event])):
                for j in range(int(values['-row_2-'])):
                    window_add_sub['2_'+str(j+1)+'_'+str(i+1)].Update(visible = True)
                    for k in range(i+1,10):
                        window_add_sub['2_'+str(j+1)+'_'+str(k+1)].Update(visible = False)
            
            # Hide answer and matrix when changing number of columns
            window_add_sub['-answer-'].Update(visible=False)
            for i in range(10):
                for j in range(10):
                    window_add_sub['3_'+str(i+1)+'_'+str(j+1)].Update(visible = False)
      
        # Show error message if M1 and M2 have a different amount of rows and/or columns 
        # (necessity for addition/subtraction)
        try:
            if values['-row_1-'] != values['-row_2-'] or values['-col_1-'] != values['-col_2-']:
                window_add_sub['-E1-'].Update(visible = True)
                for i in range(10):
                    for j in range(10):
                        window_add_sub['3_'+str(i+1)+'_'+str(j+1)].Update(visible = False)
                window_add_sub['-answer-'].Update(visible=False) 
            else:
                window_add_sub['-E1-'].Update(visible = False)
        except:
            pass
        
        # Show error message if not all values are filled in
        try:
            check_fill = 0
            for i in range(int(values['-row_1-'])):
                for j in range(int(values['-col_1-'])):
                    if values['1_'+str(i+1)+'_'+str(j+1)] == '':
                        check_fill += 1
            for k in range(int(values['-row_2-'])):
                for l in range(int(values['-col_2-'])):
                    if values['2_'+str(k+1)+'_'+str(l+1)] == '':
                        check_fill += 1
            if check_fill > 0:
                window_add_sub['-E2-'].Update(visible = True)
                for i in range(10):
                    for j in range(10):
                        window_add_sub['3_'+str(i+1)+'_'+str(j+1)].Update(visible = False)
                window_add_sub['-answer-'].Update(visible=False)  
            else:
                window_add_sub['-E2-'].Update(visible = False)
        except:
            pass

        # Show error message if not all values are floats/integers
        try:
            check_str = 0
            for i in range(int(values['-row_1-'])):
                for j in range(int(values['-col_1-'])):
                    if isfloat(values['1_'+str(i+1)+'_'+str(j+1)]) == False:
                        check_str += 1
            for k in range(int(values['-row_2-'])):
                for l in range(int(values['-col_2-'])):
                    if isfloat(values['2_'+str(k+1)+'_'+str(l+1)]) == False:
                        check_str += 1
            if check_str > 0:
                window_add_sub['-E3-'].Update(visible = True)
                for i in range(10):
                    for j in range(10):
                        window_add_sub['3_'+str(i+1)+'_'+str(j+1)].Update(visible = False)
                window_add_sub['-answer-'].Update(visible=False)  
            else:
                window_add_sub['-E3-'].Update(visible = False)
        except:
            pass
                            
        # Show addition/subtraction buttons if all values are filled in correctly
        try:
            if values['-row_1-'] == values['-row_2-'] and values['-col_1-'] == values['-col_2-'] \
            and check_fill == 0 and check_str == 0:
                window_add_sub['-add-'].Update(visible = True)
                window_add_sub['-sub-'].Update(visible = True)
            else:
                window_add_sub['-add-'].Update(visible = False)
                window_add_sub['-sub-'].Update(visible = False)
        except:
            pass
    
        # Compute addition
        if event == '-add-':
            
            # Hide answer and matrix (to prevent user from seeing too many boxes 
            # when computing a new addition with less rows and columns)
            for i in range(10):
                    for j in range(10):
                        window_add_sub['3_'+str(i+1)+'_'+str(j+1)].Update(visible = False)
            
            # Transfer values from first matrix to a list (of lists)
            M_1 = []
            for i in range(int(values['-row_1-'])):
                row = []
                for j in range(int(values['-col_1-'])):
                    while True:
                        try:
                            row.append(float(values['1_'+str(i+1)+'_'+str(j+1)]))
                            break
                        except:
                            continue
                M_1.append(row)
            
            # Transfer values from second matrix to a list (of lists)
            M_2 = []
            for i in range(int(values['-row_2-'])):
                row = []
                for j in range(int(values['-col_2-'])):
                    while True:
                        try:
                            row.append(float(values['2_'+str(i+1)+'_'+str(j+1)]))
                            break
                        except:
                            continue
                M_2.append(row)
                
            # Calculate resulting matrix    
            M_3 = [] 
            for i in range(int(values['-row_1-'])): # Defines how many rows the resulting matrix has
                row = []
                for j in range(int(values['-col_2-'])): # Defines how many columns the resulting matrix has
                    ent = 0
                    # Calculate all indices of M3 rowwise 
                    # (one M3 index equals the sum of the respective M1 and M2 indices)
                    ent += M_1[i][j] + M_2[i][j]
                    row.append(ent)
                M_3.append(row)
                
            # Show results
            for i in range(int(values['-row_1-'])):
                for j in range(int(values['-col_1-'])):
                    window_add_sub['3_'+str(i+1)+'_'+str(j+1)].Update(visible=True)
                    window_add_sub['3_'+str(i+1)+'_'+str(j+1)].Update(str(M_3[i][j]))
            window_add_sub['-answer-'].Update(visible=True) 
            
        # Compute subtraction
        if event == '-sub-':
            
            # Hide answer and matrix (to prevent user from seeing too many boxes 
            # when computing a new subtraction with less rows and columns)
            for i in range(10):
                    for j in range(10):
                        window_add_sub['3_'+str(i+1)+'_'+str(j+1)].Update(visible = False)
            
            # Transfer values from first input matrix to a list (of lists)
            M_1 = []
            for i in range(int(values['-row_1-'])):
                row = []
                for j in range(int(values['-col_1-'])):
                    while True:
                        try:
                            row.append(float(values['1_'+str(i+1)+'_'+str(j+1)]))
                            break
                        except:
                            continue
                M_1.append(row)
                
            # Transfer values from second input matrix to a list (of lists)    
            M_2 = []
            for i in range(int(values['-row_2-'])):
                row = []
                for j in range(int(values['-col_2-'])):
                    while True:
                        try:
                            row.append(float(values['2_'+str(i+1)+'_'+str(j+1)]))
                            break
                        except:
                            continue
                M_2.append(row)
                
                
            # Calculate resulting matrix    
            M_3 = [] 
            for i in range(int(values['-row_1-'])): # Defines how many rows the resulting matrix has
                row = []
                for j in range(int(values['-col_2-'])): # Defines how many columns the resulting matrix has
                    ent = 0
                    # Calculate all indices of M3 rowwise 
                    # (one M3 index equals the difference of the respective M1 and M2 indices)
                    ent += M_1[i][j] - M_2[i][j]
                    row.append(ent)
                M_3.append(row)
    
            # Show results
            for i in range(int(values['-row_1-'])):
                for j in range(int(values['-col_1-'])):
                    window_add_sub['3_'+str(i+1)+'_'+str(j+1)].Update(visible=True)
                    window_add_sub['3_'+str(i+1)+'_'+str(j+1)].Update(str(M_3[i][j]))
            window_add_sub['-answer-'].Update(visible=True)  
            
        # End if user closes window (jump back to main window)
        if event == sg.WIN_CLOSED:
            break
    
    window_add_sub.close()
    
#if __name__ == "__main__":
#    add_sub_window()
```

```python
###############################################################################################################
#####                                            INVERSE WINDOW                                           #####
###############################################################################################################

# Sub window for Inverse
def inv_window(): 
    
    # Define window for first input matrix (M1)
    matrix_1 = [
        [sg.Text('Please indicate the Dimensions of the Matrix:')],
        [
            sg.Text('Rows:'),
            sg.Slider((2,10), enable_events=True, default_value=2, \
                      size=(12,20), key='-row_1-', orientation='h'),
            sg.Text('Columns:'),
            sg.Slider((2,10), enable_events=True, default_value=2, \
                      size=(12,20), key='-col_1-', orientation='h')
        ],
        [sg.HorizontalSeparator()],
        [sg.Text('Please fill in the respective values for the Matrix:')],        
    ]
    # Add 10x10 input boxes to M1
    for i in range(10):
        box_list=[]
        for j in range(10):
            # (First four boxes should be visible from the beginning)
            if (i == 0 or i == 1) and (j == 0 or j == 1): 
                box_list.append(sg.pin(sg.In(size=(3,2), enable_events=True, \
                key = '1_'+str(i+1)+'_'+str(j+1), visible=True, justification='c')))
            else:
                box_list.append(sg.pin(sg.In(size=(3,2), enable_events=True, \
                key = '1_'+str(i+1)+'_'+str(j+1), visible=False, justification='c')))            
        matrix_1.append(box_list) 
        
    # Define window for output matrix (M3)    
    matrix_3 = []
    # Add 10x10 input boxes to M3
    for i in range(10):
        box_list=[]
        for j in range(10):
            # (In the beginning, all boxes should be invisible)
            box_list.append(sg.pin(sg.In(size=(8,2), enable_events=True, \
            key = '3_'+str(i+1)+'_'+str(j+1), visible=False, justification='c')))            
        matrix_3.append(box_list) 
        
    # Define layout for inverse window    
    layout_inv = [
                [sg.Column(matrix_1, size=(360, 380), element_justification='c')],
                [sg.HorizontalSeparator()],
                [sg.pin(sg.Button('Compute Inverse', key= '-inv-', size = (35,1), visible=False))],
                [sg.Text('The resulting Matrix is:', key = '-answer-',visible=False)],
                [sg.Column(matrix_3)],
                [sg.Text('Must be a square Matrix!', key = '-E1-', text_color='red', \
                         background_color='white', visible=False)],
                [sg.Text('Please fill in all missing values!', key = '-E2-',text_color='red', \
                         background_color='white', visible=True)],    
                [sg.Text('All Entries must be floats or integers!', key = '-E3-',text_color='red', \
                         background_color='white', visible=True)],    
                [sg.Text('The Matrix is singular and thus not invertible!', key = '-E4-',text_color='red', \
                         background_color='white', visible=False)]
    ]
    
    # Define inverse window
    window_inv = sg.Window('Matrix Calculations: Inverse', layout_inv, \
                           size=(780,740), element_justification='c', modal=True)
    
    # Event loop
    while True:
        event, values = window_inv.read()
            
        # Create (show) respective input fields for M1 rows
        if event == '-row_1-':
            for i in range(int(values[event])):
                for j in range(int(values['-col_1-'])):
                    window_inv['1_'+str(i+1)+'_'+str(j+1)].Update(visible = True)
                    for k in range(i+1,10):
                        window_inv['1_'+str(k+1)+'_'+str(j+1)].Update(visible = False)
            
            # Hide answer and matrix when changing number of rows
            window_inv['-answer-'].Update(visible=False)
            window_inv['-E4-'].Update(visible = False)
            for i in range(10):
                for j in range(10):
                    window_inv['3_'+str(i+1)+'_'+str(j+1)].Update(visible = False)
                            
        # Create (show) respective input fields for M1 columns    
        if event == '-col_1-':
            for i in range(int(values[event])):
                for j in range(int(values['-row_1-'])):
                    window_inv['1_'+str(j+1)+'_'+str(i+1)].Update(visible = True)
                    for k in range(i+1,10):
                        window_inv['1_'+str(j+1)+'_'+str(k+1)].Update(visible = False)
            
            # Hide answer and matrix when changing number of columns
            window_inv['-answer-'].Update(visible=False)
            window_inv['-E4-'].Update(visible = False)
            for i in range(10):
                for j in range(10):
                    window_inv['3_'+str(i+1)+'_'+str(j+1)].Update(visible = False)
      
        # Show error message if M1 and M2 have different dimensions (square Matrix necessity for inverse)
        try:
            if values['-row_1-'] != values['-col_1-']:
                window_inv['-E1-'].Update(visible = True)
                for i in range(10):
                    for j in range(10):
                        window_inv['3_'+str(i+1)+'_'+str(j+1)].Update(visible = False)
                window_inv['-answer-'].Update(visible=False)
                window_inv['-E4-'].Update(visible = False)
            else:
                window_inv['-E1-'].Update(visible = False)
        except:
            pass
        
        # Show error message if not all values are filled in
        try:
            check_fill = 0
            for i in range(int(values['-row_1-'])):
                for j in range(int(values['-col_1-'])):
                    if values['1_'+str(i+1)+'_'+str(j+1)] == '':
                        check_fill += 1
            if check_fill > 0:
                window_inv['-E2-'].Update(visible = True)
                for i in range(10):
                    for j in range(10):
                        window_inv['3_'+str(i+1)+'_'+str(j+1)].Update(visible = False)
                window_inv['-answer-'].Update(visible=False)
                window_inv['-E4-'].Update(visible = False)
            else:
                window_inv['-E2-'].Update(visible = False)
        except:
            pass

        # Show error message if not all values are floats/integers
        try:
            check_str = 0
            for i in range(int(values['-row_1-'])):
                for j in range(int(values['-col_1-'])):
                    if isfloat(values['1_'+str(i+1)+'_'+str(j+1)]) == False:
                        check_str += 1
            if check_str > 0:
                window_inv['-E3-'].Update(visible = True)
                for i in range(10):
                    for j in range(10):
                        window_inv['3_'+str(i+1)+'_'+str(j+1)].Update(visible = False)
                window_inv['-answer-'].Update(visible=False)
                window_inv['-E4-'].Update(visible = False)
            else:
                window_inv['-E3-'].Update(visible = False)
        except:
            pass
                            
        # Show inverse button if all values are filled in correctly
        try:
            if values['-row_1-'] == values['-col_1-'] and check_fill == 0 and check_str == 0:
                window_inv['-inv-'].Update(visible = True)
            else:
                window_inv['-inv-'].Update(visible = False)
        except:
            pass
                
        # Compute inverse
        if event == '-inv-':
            
            # Hide answer and matrix (to prevent user from seeing too many boxes 
            # when computing a new inverse with less rows and columns)
            window_inv['-E4-'].Update(visible = False)
            for i in range(10):
                    for j in range(10):
                        window_inv['3_'+str(i+1)+'_'+str(j+1)].Update(visible = False)
            
            # Transfer values from input matrix to a list (of lists)
            M_1 = []
            for i in range(int(values['-row_1-'])):
                row = []
                for j in range(int(values['-col_1-'])):
                    while True:
                        try:
                            row.append(float(values['1_'+str(i+1)+'_'+str(j+1)]))
                            break
                        except:
                            continue
                M_1.append(row)
        
            # Test if determinant equals zero (then no inverse exists for this matrix)
            # Transform matrix to numpy array
            N = np.asarray(M_1, dtype=float)
            
            # Transform our matrix to the Echelon form (while loop as long as we do not have the Echelon form)
            while True:
                change = 0 # defines the times of switching rows
                for o in range(int(values['-col_1-'])):
                    for p in range(int(values['-row_1-'])):
                        if o == p: # for every entry on the diagonal axis (1:1, 2:2, ...)...
                            if N[p][o] == 0: # if the entry equals zero ...
                                for q in range(int(values['-row_1-'])): # loop through the rows and find ...
                                    if N[q][o] != 0: # a row where the entry does not equal zero to ...
                                        N[[q,p]] = N[[p,q]] # switch these rows
                                        change += 1 #(and add 1 to the change-counter)
                        if N[p][o] != 0 and p > o: # for all values under the diagonal axis ...
                            N[p] = N[p] - N[o]*(N[p][o]/N[o][o]) # subtract from that respective row ...
                                                                 # another row such that the remaining ...
                                                                 # value equals zero
                    test_1 = 0
                    test_2 = []
                    det = 1
                    for r in range(int(values['-row_1-'])):
                        for s in range(int(values['-col_1-'])):
                            if r == s:
                                det *= N[r][s]
                            if r > s:
                                test_1 += N[r][s]
                                test_2.append(N[r][s])
                    if test_1 == 0 and len(set(test_2)) == 1: # test_1 equals the sum of all values under 
                                                              # the diagonal; test_2 equals the sum of all 
                                                              # different values under the diagonal
                                                              # if both are zero, we have the Echelon form
                        break
                if test_1 == 0 and len(set(test_2)) == 1:
                    break
    
            det = (-1)**change*det            
            

            if det == 0:
                window_inv['-E4-'].Update(visible = True)
            else:
        
                # Append identity matrix
                for m in range(int(values['-row_1-'])):
                    for n in range(int(values['-col_1-'])):
                        if m == n:
                            M_1[m].append(1)
                        else:
                            M_1[m].append(0)
                
                M_3 = np.asarray(M_1, dtype=float)
                
                # Use the logarithm learned in class to transform original matrix to identity matrix
                # (while loop as long as we do not have the identity matrix 'on the left side')
                while True:
                    for o in range(int(values['-col_1-'])): ## for each column of the Matrix
                        for p in range(int(values['-row_1-'])): ## for each row of the Matrix
                            if o == p: ## for every entry on the diagonal axis (1:1, 2:2, ...)
                                if M_3[p][o] == 0: 
                                ## if entry on diagonal axis is zero, row should be switched with
                                ## another row which does not have a zero entry for that column    
                                    ## find the suitable row in the Matrix
                                    for q in range(int(values['-row_1-'])): 
                                        ## row only suitable if entry for that column is not zero
                                        if M_3[q][o] != 0: 
                                            M_3[[q,p]] = M_3[[p,q]] ## switch respective rows
                                ## (after switching) divide row by the value of its entry 
                                ## on the diagonal axis
                                M_3[p] = M_3[p]/M_3[p][o] 
                            else: ## for every entry NOT on the diagonal axis ... 
                                if M_3[p][o] != 0: ## which does not equal to zero ... 
                                    M_3[p] = M_3[p] - M_3[o]*M_3[p][o] ## 
                                #print(N)
                    test_1 = 0
                    test_2 = 0
                    for r in range(int(values['-row_1-'])):
                        test_1 += sum(M_3[r][:int(values['-row_1-'])])
                        test_2 += len(set(M_3[r][:int(values['-row_1-'])]))
                    if test_1 == int(values['-row_1-']) and test_2 == 2*int(values['-row_1-']):
                    # test_1 equals the sum of all values of all rows (of the 'left' matrix)
                    # test_2 equals the sum of the sum of all different values per row (of the 'left' matrix)
                    # If test_1 equals the number of rows/columns and test_2 equals twice the number of 
                    # rows/columns (cause there should be only two different numbers per row: 0 and 1),
                    # we have the identity matrix
                        break          
                        
                # Show results
                for i in range(int(values['-row_1-'])):
                    for j in range(int(values['-col_1-'])):
                        window_inv['3_'+str(i+1)+'_'+str(j+1)].Update(visible=True)
                        window_inv['3_'+str(i+1)+'_'+str(j+1)]\
                        .Update(str(M_3[i][j+int(values['-row_1-'])]))
                window_inv['-answer-'].Update(visible=True)  
            
        # End if user closes window (jump back to main window)
        if event == sg.WIN_CLOSED:
            break
    
    window_inv.close()
    
#if __name__ == "__main__":
#    inv_window()
```
